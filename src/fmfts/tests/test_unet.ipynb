```python
import torch
from fmfts.utils.unet import UNet

# tests a 3d UNet
unet = UNet(
    in_channels=4, out_channels=5,
    features = (32, 64, 96),
    padding = ("zeros", "replicate", "circular"),
    nl = torch.nn.ReLU()
) 
x = torch.randn(1, 4, 16, 16, 16)
print( unet(x).shape )


# tests a 2d UNet
unet = UNet(
    in_channels=4, out_channels=5,
    features = (32, 64, 96),
    padding = ("zeros", "replicate"),
    nl = torch.nn.ReLU()
) 
x = torch.randn(1, 4, 16, 16)
print( unet(x).shape )


# tests a 1d UNet
unet = UNet(
    in_channels=4, out_channels=5,
    features = (32, 64, 96),
    padding = ("zeros",),
    nl = torch.nn.ReLU()
) 
x = torch.randn(1, 4, 16)
print( unet(x).shape )
```
