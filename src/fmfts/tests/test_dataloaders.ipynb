```python
import matplotlib.pyplot as plt
```

```python
from fmfts.dataloader.ks2d import DatasetKS2D

dataset = DatasetKS2D(mode="train")

y, x = dataset[0]
fig, ax = plt.subplots(1, 2, figsize=(3*2, 3))

ax[0].set_title("x(t)")
ax[1].set_title("x(t+dt)")
dataset.plot(y, ax[0])
dataset.plot(x, ax[1])
```

```python
from fmfts.dataloader.ns2d import DatasetNS2D

dataset = DatasetNS2D(mode="train")

y, x = dataset[0]
fig, ax = plt.subplots(3, 2, figsize=(3*2, 3*3))
ax[0,0].set_title("x(t)")
ax[0,1].set_title("x(t+dt)")
for i, viz in enumerate(["sqrt_energy", "speed", "streamlines"]):
    dataset.plot(y, ax[i,0], visualization=viz)
    dataset.plot(x, ax[i,1], visualization=viz)
    ax[i,0].set_ylabel(viz)
```

```python
from fmfts.dataloader.rti3d_full import DatasetFullRTI3D

dataset = DatasetFullRTI3D(At=3/4, mode="train", dt=10, dx=2, dy=2, dz=2)

y, x = dataset[500]
fig, ax = plt.subplots(1, 2, figsize=(3*2, 3), subplot_kw=dict(projection="3d"))
dataset.plot(y, ax[0])
dataset.plot(x, ax[1])
ax[0].set_title("x(t)")
ax[1].set_title("x(t+dt)")
plt.show()
```

```python
from fmfts.dataloader.rti3d_sliced import DatasetSlicedRTI3D

dataset = DatasetSlicedRTI3D(At=3/4, mode="train", dt=10, dy=1, dz=1)

y, x = dataset[500]
fig, ax = plt.subplots(1, 2, figsize=(3*2, 3))
dataset.plot(y, ax[0])
dataset.plot(x, ax[1])
ax[0].set_title("x(t)")
ax[1].set_title("x(t+dt)")
plt.show()
```
